```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
browser = webdriver.Firefox()
```

```python
browser.get(url)
```

```python
html = browser.page_source
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
data=[]
for r in soup.select('table.wikitable.sortable.jquery-tablesorter')[1].tbody.find_all('tr'):
    data.append([i.text.strip() for i in r.find_all('td')])
```

```python
def bomb_table():
    thead = [i.text.strip() for i in table2.find_all('th')]
    data=[]
    for r in soup.select('table.wikitable.sortable.jquery-tablesorter')[1].tbody.find_all('tr'):
        data.append([i.text.strip() for i in r.find_all('td')])
    df = pd.DataFrame(data)
    df.columns = thead
    return df
```

```python
bomb_table()
```

```python
def bomb_all(table_no):
    thead = [i.text.strip() for i in table2.find_all('th')]
    data=[]
    for r in table_no.tbody.find_all('tr'):
        data.append([i.text.strip() for i in r.find_all('td')])
    df = pd.DataFrame(data)
    df.columns = thead
    return df
```
